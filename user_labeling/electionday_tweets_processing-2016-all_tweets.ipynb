```python
import pandas
import numpy as np

df = pandas.read_csv('2016_allStates_tweets.csv', encoding='utf8', engine='python')
df
```

```python
new = df["name"].str.split(" ", expand = True)

df["user_fn"]= new[0] 
df["user_ln"]= new[1] 

df
```

```python
df_race = pandas.read_csv('condensed_lastnames.csv', header=None)
df_race.columns = ["last_names", "race/ethnicity"]
df_race
```

```python
df["user_ln"] = df["user_ln"].str.lower()
df_tweets_ln = df.set_index('user_ln')

df_race["last_names"] = df_race["last_names"].str.lower()
df_race_ln = df_race.set_index('last_names')

df_tweets_ln[['race/ethnicity']] = df_race_ln[['race/ethnicity']]
df_tweets_ln
```

```python
df_with_race = df_tweets_ln.reset_index()
df_with_race
```

```python
df_with_race.groupby('race/ethnicity').count()
```

```python
df_gender = pandas.read_csv('condensed_first_names.csv', header=None)
df_gender.columns = ['firstnames', 'gender', 'count', 'grouped', 'grouped_percent']
df_gender.drop(['count', 'grouped', 'grouped_percent'], axis = 1, inplace = True)
df_gender
```

```python
df_with_race["user_fn"] = df_with_race["user_fn"].str.lower()
df_tweets_fn = df_with_race.set_index('user_fn')
display(df_tweets_fn)

df_gender["firstnames"] = df_gender["firstnames"].str.lower()
df_gender_fn = df_gender.set_index('firstnames')
display(df_gender_fn)

df_tweets_fn[['gender']] = df_gender_fn[['gender']]
```

```python
df_tweets_fn
```

```python
df_race_and_gender = df_tweets_fn.reset_index()
```

```python
df_race_and_gender
```

```python
df_race_and_gender.drop(['urls', 'photos', 'video', 'thumbnail'], axis = 1, inplace = True)
df_race_and_gender.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1, inplace = True)
df_race_and_gender
```

```python
df_race_and_gender.groupby('race/ethnicity').count()
```

```python
df_race_and_gender.groupby('gender').count()
```

```python
import csv
df_race_and_gender.to_csv('race&gender-2016-all.csv',quoting=csv.QUOTE_ALL)
```
