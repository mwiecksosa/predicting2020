```python
import pandas
import numpy as np
df = pandas.read_csv('NationalNames.csv')
df.drop(['Year', 'Id'], axis = 1, inplace = True) 
print(df.head())
```

```python
grouped = df.groupby(['Name', 'Gender'])['Count'].sum()
df['grouped'] = df.groupby(['Name', 'Gender'])['Count'].transform('sum')
df = df.sort_values("Name", ascending = False)
df = df.drop_duplicates(["Name", "Gender"])

grouped_percent = df['grouped']/df.groupby(['Name'])['grouped'].sum()
df['grouped_percent'] = df['grouped']/df.groupby('Name')['grouped'].transform('sum')
df = df.sort_values("grouped", ascending = False)
df
```

```python
df2 = df[df['grouped_percent'] > .80] 

df2
```

```python
np.savetxt("condensed_first_names.csv", df2, fmt='%s', delimiter=",")
```
