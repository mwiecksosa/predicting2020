```python
import pandas
import numpy as np
df = pandas.read_csv('2016_formattedForWebsite.csv')
df
```

```python
pandas.set_option('display.max_columns', None)
df.head()
```

```python
df_politics = df.filter(['democratCounter','republicanCounter','neutralCounter','unknownPoliticsCounter', 'state_full'], axis=1)
```

```python
conditions = [
    (df_politics['neutralCounter'] > df_politics['democratCounter']) & (df_politics['neutralCounter'] > df_politics['republicanCounter']),
    (df_politics['democratCounter'] > df_politics['republicanCounter']),
    (df_politics['republicanCounter'] > df_politics['democratCounter']),
    
    ]

# create a list of the values we want to assign for each condition
values = ['N', 'D', 'R']

# create a new column and use np.select to assign values to it using our lists as arguments
df_politics['state_vote'] = np.select(conditions, values)

# display updated DataFrame
df_politics
```

```python
df_gender = df.filter(['womanDemocratCounter','womanRepublicanCounter','womanNeutralCounter', 'manDemocratCounter','manRepublicanCounter','manNeutralCounter', 'state_full'], axis=1)
df_gender['totalWomen'] = df_gender['womanDemocratCounter'] + df_gender['womanRepublicanCounter'] + df_gender['womanNeutralCounter']
df_gender['totalMen'] = df_gender['manDemocratCounter'] + df_gender['manRepublicanCounter'] + df_gender['manNeutralCounter']
df_gender
```

```python
df_gender['correctedDemocratCounter'] = (((df_gender['womanDemocratCounter']/df_gender['totalWomen']))*.55 + ((df_gender['manDemocratCounter']/df_gender['totalMen']))*.45)*(df_gender['totalWomen']+df_gender['totalMen'])
```

```python
df_gender['correctedRepublicanCounter'] = (((df_gender['womanRepublicanCounter']/df_gender['totalWomen']))*.55 + ((df_gender['manRepublicanCounter']/df_gender['totalMen']))*.45)*(df_gender['totalWomen']+df_gender['totalMen'])
df_gender['correctedNeutralCounter'] = (((df_gender['womanNeutralCounter']/df_gender['totalWomen']))*.55 + ((df_gender['manNeutralCounter']/df_gender['totalMen']))*.45)*(df_gender['totalWomen']+df_gender['totalMen'])
df_gender
df_gender_results = df_gender.filter(['correctedDemocratCounter', 'correctedRepublicanCounter', 'correctedNeutralCounter', 'state_full'], axis = 1 )
```

```python
df_gender_results
```

```python
conditions = [
    (df_gender_results['correctedNeutralCounter'] > df_gender_results['correctedDemocratCounter']) & (df_gender_results['correctedNeutralCounter'] > df_gender_results['correctedRepublicanCounter']),
    (df_gender_results['correctedDemocratCounter'] > df_gender_results['correctedRepublicanCounter']),
    (df_gender_results['correctedRepublicanCounter'] > df_gender_results['correctedDemocratCounter']),
    
    ]

# create a list of the values we want to assign for each condition
values = ['N', 'D', 'R']

# create a new column and use np.select to assign values to it using our lists as arguments
df_gender_results['state_vote_corrected'] = np.select(conditions, values)

# display updated DataFrame
df_gender_results

import csv
df_gender_results.to_csv('simple_gender_corrected_2016.csv')
```

```python
df_race = df.filter(['aianDemocratCounter','aianRepublicanCounter','aianNeutralCounter', 'apiDemocratCounter','apiRepublicanCounter','apiNeutralCounter', 'blackDemocratCounter','blackRepublicanCounter','blackNeutralCounter', 'whiteDemocratCounter','whiteRepublicanCounter','whiteNeutralCounter', 'hispanicDemocratCounter','hispanicRepublicanCounter','hispanicNeutralCounter', 'state_full'], axis=1)
df_race['otherDemocratCounter'] = df_race['aianDemocratCounter'] + df_race['apiDemocratCounter']
df_race['otherRepublicanCounter'] = df_race['aianRepublicanCounter'] + df_race['apiRepublicanCounter']
df_race['otherNeutralCounter'] = df_race['aianNeutralCounter'] + df_race['apiNeutralCounter']
df_race
```

```python
df_race['totalOther'] = df_race['otherDemocratCounter'] + df_race['otherRepublicanCounter'] + df_race['otherNeutralCounter']
df_race['totalWhite'] = df_race['whiteDemocratCounter'] + df_race['whiteRepublicanCounter'] + df_race['whiteNeutralCounter']
df_race['totalBlack'] = df_race['blackDemocratCounter'] + df_race['blackRepublicanCounter'] + df_race['blackNeutralCounter']
df_race['totalHispanic'] = df_race['hispanicDemocratCounter'] + df_race['hispanicRepublicanCounter'] + df_race['hispanicNeutralCounter']
df_race
```

```python
df_race['totalBlack'] = df_race['totalBlack'] + .000001
df_race['totalOther'] = df_race['totalOther'] + .000001
df_race['totalHispanic'] = df_race['totalHispanic'] + .000001

df_race['correctedDemocratCounter'] = (((df_race['whiteDemocratCounter']/df_race['totalWhite']))*.74 + ((df_race['blackDemocratCounter']/df_race['totalBlack']))*.1 + ((df_race['otherDemocratCounter']/df_race['totalOther']))*.05 + ((df_race['hispanicDemocratCounter']/df_race['totalHispanic']))*.1)*(df_race['totalOther']+df_race['totalWhite'] + df_race['totalBlack'] + df_race['totalHispanic'])
df_race['correctedRepublicanCounter'] = (((df_race['whiteRepublicanCounter']/df_race['totalWhite']))*.74 + ((df_race['blackRepublicanCounter']/df_race['totalBlack']))*.1 + ((df_race['otherRepublicanCounter']/df_race['totalOther']))*.05 + ((df_race['hispanicRepublicanCounter']/df_race['totalHispanic']))*.1)*(df_race['totalOther']+df_race['totalWhite'] + df_race['totalBlack'] + df_race['totalHispanic'])
df_race['correctedNeutralCounter'] = (((df_race['whiteNeutralCounter']/df_race['totalWhite']))*.74 + ((df_race['blackNeutralCounter']/df_race['totalBlack']))*.1 + ((df_race['otherNeutralCounter']/df_race['totalOther']))*.05 + ((df_race['hispanicNeutralCounter']/df_race['totalHispanic']))*.1)*(df_race['totalOther']+df_race['totalWhite'] + df_race['totalBlack'] + df_race['totalHispanic'])

df_race_results = df_race.filter(['correctedDemocratCounter', 'correctedRepublicanCounter', 'correctedNeutralCounter', 'state_full'], axis = 1 )

df_race_results


```

```python
conditions = [
    (df_race_results['correctedNeutralCounter'] > df_race_results['correctedDemocratCounter']) & (df_race_results['correctedNeutralCounter'] > df_race_results['correctedRepublicanCounter']),
    (df_race_results['correctedDemocratCounter'] > df_race_results['correctedRepublicanCounter']),
    (df_race_results['correctedRepublicanCounter'] > df_race_results['correctedDemocratCounter']) 
    ]

# create a list of the values we want to assign for each condition
values = ['N', 'D', 'R']

# create a new column and use np.select to assign values to it using our lists as arguments
df_race_results['state_vote_corrected'] = np.select(conditions, values)

# display updated DataFrame
df_race_results

df_race_results.to_csv('simple_race_corrected_2016.csv')
```
